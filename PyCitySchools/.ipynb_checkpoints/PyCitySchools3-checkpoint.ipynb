```python
# Dependencies and Setup
import pandas as pd
import locale

locale.setlocale(locale.LC_ALL, '')

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
df = school_data_complete
```

```python
# Find number of schools
total_schools = len(df['School ID'].unique())
total_schools
```

```python
# Find number of students
total_students = df['Student ID'].count()
total_students = f'{total_students:n}'
```

```python
#Calculate total budget
total_budget = df['budget'].unique().sum()
total_budget_f = f'${total_budget:n}'
print(total_budget_f)
```

```python
# Find average of math scores
avg_math = df['math_score'].mean()
avg_math
```

```python
# Find average reading scores
avg_reading = df['reading_score'].mean()
avg_reading
```

```python
# Find percentage of students who scored over or equal to 70: math
pass_math = (df['math_score']>=70).mean()
pass_math
```

```python
# Find percentage of students who scored over or equal to 70: reading
pass_reading = (df['reading_score']>=70).mean()
pass_reading
```

```python
# Find percentage that passed overall
overall_pass = (pass_math + pass_reading)/2
overall_pass
```

```python
# Create dictionary with variable values
district_sum = {
    "Total Schools": total_schools,
    "Total Students": total_students,
    "Total Budget": total_budget_f,
    "Average Math Score": avg_math,
    "Average Reading Score": avg_reading,
    "%Passing Math": pass_math,
    "%Passing Reading": pass_reading,
    "%Overall Passing Rate": overall_pass
}
```

```python
# Turn into dataframe
district_sum_df = pd.DataFrame(district_sum, index=[0])
```

```python
district_sum_df
```

```python
df['passes_math'] = df['math_score'] >=70
```

```python
df['passes_reading'] = df['reading_score'] >= 70
```

```python
# Grouping by school
df_school = df.groupby(['school_name']).mean()
```

```python
df_school
```

```python
# Calculating needed variables
df_school["Per Student Budget"] = df_school['budget']/df_school['size']
```

```python
df_school['% Passing Math'] = df_school['passes_math']*100
```

```python
df_school['% Passing Reading'] =df_school['passes_reading']*100
```

```python
df_school['% Overall Passing'] = (df_school['% Passing Math'] + df_school['% Passing Reading'])/2
```

```python
# Merging to add type, size, and budget
school_summ = pd.merge(df_school, school_data, how='left', on=["school_name", "school_name"])

```

```python
df_school
```

```python
del school_summ['size_y']
del school_summ['budget_y']
del school_summ['Student ID']
del school_summ['School ID_x']
```

```python
# Create dataframe for school summary info
school_summ_dataframe = pd.DataFrame({
    "School Name": school_summ["school_name"],
    "School Type": school_summ["type"],
    "Total Students": school_summ["size_x"].apply(lambda x: "{:.0f}".format(x)),
    "Total School Budget": school_summ['budget_x'].apply(lambda x: "${:,.2f}".format(x)),
    "Per Student Budget": school_summ['Per Student Budget'].apply(lambda x: "${:,.2f}".format(x)),
    "Average Math Score": school_summ['math_score'],
    "Average Reading Score": school_summ['reading_score'],
    "% Passing Math": school_summ['passes_math'],
    "% Passing Reading": school_summ['passes_reading'],
    "% Overall Passing Rate": school_summ['% Overall Passing']                                      
})
```

```python
# Format decimals
school_summ_dataframe['% Passing Math'] = school_summ_dataframe['% Passing Math']*100
school_summ_dataframe['% Passing Reading'] = school_summ_dataframe['% Passing Reading']*100

school_summ_dataframe
```

```python
# Sort for top 5 based on overall passing percentage
school_summ_dataframe.sort_values('% Overall Passing Rate',ascending =False).head(5)
```

```python
# Sort for bottom 5 based on overall passing percentage
school_summ_dataframe.sort_values('% Overall Passing Rate',ascending =False).tail(5)
```

```python
# Scores by School Spending

spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
```

```python
school_summ["Per Student Budget(Ranges)"] = pd.cut(school_summ["Per Student Budget"], spending_bins, labels=group_names)
```

```python
school_spend_gp = school_summ.groupby("Per Student Budget(Ranges)").mean()
```

```python
# Delete irrelevant columns
del school_spend_gp['size_x']
del school_spend_gp['budget_x']
del school_spend_gp['Per Student Budget']
del school_spend_gp['School ID_y']
del school_spend_gp['passes_math']
del school_spend_gp['passes_reading']
```

```python
school_spend_gp = school_spend_gp.rename(columns = {
    'reading_score': 'Average Reading Score',
    'math_score': 'Average Math Score'
})
```

```python
school_spend_gp
```

```python
#Scores by School Size

size_bins = [0, 1000, 2000, 5000]
group_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

## Two Observable Trends

1. According to the school size table we observe that as school size increases the passing rates of both math, reading, and overall lower considerably. Large school sizes may lead to larger classrooms thud meaning students get less one on one time with teachers which may explain the difference. Average class size for each school would be something interesting to look at.

2. Another interesting observation is the differences of the top five and bottom five of the high schools. If you examin the average reading and math scores you notice the top five schools aare doing better but not by much, where the huge gap lies is in the % Passing. The top schools have an almost 20% higher passing in both math and reading. This leads me to think the bottom schools may have a distribution of scores in the shape of an inverted bell curve. With many low and many high scores this balances the mean but leads to the lower observable passing rate.
